### Parquet files Analysis

each file contains 100K domains

**filters:** 

    - ip_country=="UA"
    - html_country=="UA"
    
For the domains with status code = 200 the html title is converted in english

### Data Loading

```python
files = glob.glob(root+"*.parquet_title_en.csv")
for filename in files:
    print (filename)
```

```python
def readcsv(args):
    return pd.read_csv(args,quotechar='"',sep ='\t')

dt = pd.concat(
    map(readcsv, glob.glob(root+"*.parquet_title_en.csv")),ignore_index=True)
```

```python
#filename="xbz.parquet_title_en.csv"
#dt = pd.read_csv("/home/ndrberna/data/"+filename, on_bad_lines='skip',quotechar='"',sep ='\t')

dt['response_code'] = dt['response_code'].fillna(0)
dt['size'] = dt['size'].fillna(0) 
dt['html_title'] = dt['html_title'].fillna("") 
dt['guessed_language'] = dt['guessed_language'].fillna("") 
dt['title_translate'] = dt['title_translate'].fillna("").apply(lambda x: x.lower())
dt['response_code'] = dt['response_code'].astype("int32")
dt=dt[(dt['response_code']==200)][["url", "title_translate","html_title","guessed_language","size","response_code"]]
dt.head(10)
```

```python
print ("Zonefiles: ",len (files))
print ("Domains: ",dt.shape[0])
```

### Searching for Sucho keywords of interest

```python
result = pd.DataFrame()
keywords = ["edu", "school","museum","art","collection","national","holocaust","repository","geographic","ethnographic","history","tour","virtual","academy","sciences","gallery","archive","library","scientific","journal"]

display(Markdown("Zonefiles analyzed: **"+str(len(files))+"**"))
display(Markdown("Domains extracted: **"+str(dt.shape[0])+"**"))
display(Markdown("Searched keywords: **"+str(keywords)+"**"))


for i in keywords:
    display(Markdown("keyword: **"+i+"**"))
    
    title_for_keywords=dt[dt.title_translate.str.contains(i)][["url","title_translate","html_title"]]
    url_for_keywords=dt[dt.url.str.contains(i)][["url","title_translate","html_title"]]
    display(title_for_keywords)
    display(url_for_keywords)
    display(Markdown("___"))
    result=pd.concat([result,title_for_keywords,url_for_keywords])
    
    
```

```python
result= result.drop_duplicates()
result.shape
result.rename(columns = {'title_translate':'English translation', 'html_title':'Original Title'}, inplace = True)
```
