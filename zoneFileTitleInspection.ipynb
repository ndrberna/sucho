### Parquet files Analysis

For the domains with status code = 200 the html title is converted in english

### Data Loading

```python
files = glob.glob(root+"*.parquet_title_en.csv")
for filename in files:
    print (filename)
```

```python
def readcsv(args):
    return pd.read_csv(args,quotechar='"',sep ='\t')

dt = pd.concat(
    map(readcsv, glob.glob(root+"*.parquet_title_en.csv")),ignore_index=True)
```

```python
#filename="xbz.parquet_title_en.csv"
#dt = pd.read_csv("/home/ndrberna/data/"+filename, on_bad_lines='skip',quotechar='"',sep ='\t')

dt['response_code'] = dt['response_code'].fillna(0)
dt['size'] = dt['size'].fillna(0) 
dt['html_title'] = dt['html_title'].fillna("") 
dt['guessed_language'] = dt['guessed_language'].fillna("") 
dt['title_translate'] = dt['title_translate'].fillna("").apply(lambda x: x.lower())
dt['response_code'] = dt['response_code'].astype("int32")
dt=dt[(dt['response_code']==200)][["url", "title_translate","html_title","guessed_language","size","response_code"]]
dt.head(10)
```
