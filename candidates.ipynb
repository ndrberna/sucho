```python
dt.response_code = dt.response_code.fillna(0)
dt['response_code'] = dt['response_code'].astype("int32")

dt["size"] = dt["size"].fillna(0) 
dt.html_title = dt.html_title.fillna("") 
dt.guessed_language = dt.guessed_language.fillna("") 
dt.title_translate = dt.title_translate.fillna("").apply(lambda x: x.lower())
dt=dt[(dt['response_code']==200) & (dt['html_title']!="")][["url", "title_translate","html_title","guessed_language","size","response_code"]]
dt.head(10)  
dt=dt[["url","title_translate","html_title"]]
```

```python
blacklist=["shop","store","computer","buzz","xn-","the subscription plan has expired","coming soon page","parking domain","oops","wrong","massage","sex","seo","repair","sea tours","last minute ","accommodation","hot tours","last minute tours","visas","hotel","domain owner","domain parked","cheap","shopping","buy","rental","escort","dealer","sell","casino","under construction"]
for i in blacklist:
    dt=dt[~dt.title_translate.str.contains(i)]
    print (i, dt.shape)
    dt=dt[~dt.url.str.contains(i)]
    print (i, dt.shape)
```

```python
from IPython.display import Markdown, display
result = pd.DataFrame()

keywords = ["temple","science", "academy","anthology","publication","heritage","gallery","archiv","librar","scientific","journal","education", "school","institut","college","lyceum","academy","catholic","ortodox","jewish", "muslim","mosque","seminary","church", "cathedral","opera","music","theater","academic","culture","museum","art","collection","repository",
"geograph","ethnographic","histor","edu","tour","virtual","cinema","memorial","festival"]

extra_keywords=["lib","cbs","bk","teatr"]
display(Markdown("Domains extracted: **"+str(dt.shape[0])+"**"))
display(Markdown("Searched keywords: **"+str(keywords)+"**"))
display(Markdown("___"))
for i in keywords:
    display(Markdown("KEYWORD: **'"+i+"'**"))
    
    title_for_keywords=dt[dt.title_translate.str.contains(i)][["url","title_translate","html_title"]].sort_values(by=['url'])
    title_for_keywords["keyword"]=i
    url_for_keywords=dt[dt.url.str.contains(i)][["url","title_translate","html_title"]].sort_values(by=['url'])
    url_for_keywords["keyword"]=i
    display(Markdown("occurrencies of **'"+i+"'** in the html title"))
    display(title_for_keywords)
    display(Markdown("occurrencies in URL **'"+i+"'** in the url"))
    display(url_for_keywords)
    display(Markdown("___"))
    result=pd.concat([result,title_for_keywords,url_for_keywords])   
    
```

```python
result= result.drop_duplicates()
#result=result[result["title_translate"]!=""]

print("Filtered domains:",result.shape[0])
```
